```python
religious_keywords = []
for key, word_list in religious_list.items():
    religious_keywords.extend(word_list)

religious_keywords = [w.lower() for w in religious_keywords]
```

```python
conspiracy_keywords = []

for key, word_list in conspirational_list.items():
    conspiracy_keywords.extend(word_list)

conspiracy_keywords = [w.lower() for w in conspiracy_keywords]
```

```python
#καθαρίζω το text
import re

def clean_text(status_text):
    return re.sub(r'[^a-zA-Zα-ωΑ-Ω0-9\s]', '', str(status_text).lower())
```

```python
#κατηγοριοποιώ τα text, βάση του λεξικού
```

```python
def classify_narrative(text):
    text_clean = clean_text(status_text=text)

    if any(word in text_clean for word in conspiracy_keywords):
        return "Conspiracy"
    elif any(word in text_clean for word in religious_keywords):
        return "Religious"
    else:
        return "Unknown"
```

```python
ts_df['narrative_type'] = ts_df['status_text'].apply(classify_narrative)
```

#Bert-based Semantic Retrieval


```python
#Φτιάχνω ένα αντίγραφο df για να μην χαλάσω το αρχικό
df = ts_df.copy()
```

```python
df = df.rename(columns={"status_text": "text"})
```

```python
#καθαρισμός της στήλης df για να βεβαιωθώ ότι δεν υπάρχουν κενά
df = df.dropna(subset=["text"])
df["text"] = df["text"].astype(str)
df = df[df["text"].str.len() > 5]
```

```python
#ορισμός seeds και candidate pool, Θέλω να ξεχωρίσω: 1)τα ήδη γνωστά Religious texts 2)τα Unknown texts όπου θα ψάξω νέα παραδείγματα
religious_df = df[df["narrative_type"] == "Religious"]
unknown_df = df[df["narrative_type"] == "Unknown"]
print("Religious:", len(religious_df))
print("Unknown:", len(unknown_df))
```

```python
#επιλογή embedding model
!pip install chromadb sentence-transformers pandas
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
```

```python
#Δημιουργία Embeddings, μετατρέπω τα text σε vectors
candidate_texts = unknown_df["text"].tolist()
candidate_ids = unknown_df.index.astype(str).tolist() #Για να ξέρω μετά ποιο embedding αντιστοιχεί σε ποιο κείμενο.
```

```python
candidate_embeddings = model.encode(candidate_texts, show_progress_bar=True) #Έναν πίνακα όπου:1)κάθε γραμμή αντιστοιχεί σε ένα text και 2)κάθε γραμμή είναι ένα embedding vector
```

```python
#Δημιουργία Chroma Collection (ένας πίνακας μέσα στη Chroma, όπου αποθηκεύω:ids,documents,embeddings), φτιάχνω τη vector βάση μου
import chromadb

client = chromadb.Client()
collection = client.create_collection(name="religious_candidates")
```

```python
#Αποθήκευση embeddings στη Chroma, αποθηκέυεται το μοναδικό id, το πραγματικό text, και το embedding, δηλαδή το vector represantation του text
collection.add(
    ids=candidate_ids,
    documents=candidate_texts,
    embeddings=candidate_embeddings
)
```

```python
#ορισμός seed queries, παίρνω τα ήδη γνωστά religious texts,
seed_texts = religious_df["text"].tolist()
```

```python
#μετατρέπω κάθε seed σε embedding, για κάθε Religious text πρέπει να κάνω embedding, ώστε να ψάξω ποια candidate texts είναι κοντά του
#Αποθήκευση αποτελεσμάτων
#Για να συγκεντρώσω τα αποτελέσματα χρειάζομαι μια λίστα
results = []

for query_text in seed_texts:
    query_embedding = model.encode([query_text])

    res = collection.query(
        query_embeddings=query_embedding,
        n_results=10
    )

    for i in range(len(res["ids"][0])):
        results.append({
            "query_text": query_text,
            "matched_id": res["ids"][0][i],
            "matched_text": res["documents"][0][i],
            "distance": res["distances"][0][i]
        })
query_embedding = model.encode([query_text])
```

```python
#similarity search στη chroma
res = collection.query(
    query_embeddings=query_embedding,
    n_results=10                       #τα 10 πιο κοντινά embeddings μέσα στη βάση, ποια unknown texts είναι πιο κοντά,τι ακριβώς γράφουν, πόσο μεγάλη είναι η απόστασh
)
```

```python
#μετατροπή σε dataframe
results_df = pd.DataFrame(results)
results_df
```

```python
#βλέπω ότι έχω πολλά duplicates, τα αφαιρώ
unique_results = results_df.drop_duplicates(subset=["matched_id"])
unique_results
```

```python
#Θέλω να βρω αυτα που είναι πιο κοντα, distance treshold και να αφαιρέσω τον θόρυβο, θα βάλω ένα treshold 10
filtered_results = results_df[results_df["distance"] < 10]
len(filtered_results)
```

```python
#ranking με βάση τη συχνότητα, πόσες φορές το ίδιο candidate κείμενο επεστράφη ως κοντινό σε διαφορετικά Religious seeds.
ranked = filtered_results["matched_id"].value_counts().reset_index()
ranked.columns = ["matched_id", "frequency"]
```
